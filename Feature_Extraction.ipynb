```python
import cv2
import matplotlib.pyplot as plt

from google.colab.patches import cv2_imshow
```

```python
cv2.__version__
```

##Preparaci√≥n de Imagen

```python
#Getting the reference image ready for feature detection
input_image = cv2.imread('/content/drive/MyDrive/Dive in Learning/Realidad Aumentada/BaÃÅsico/Colab Notebooks/Robot.jpg')
cv2_imshow(input_image)
#print(input_image)
```

```python
input_image.shape
```

```python
input_image = cv2.resize(input_image, (700,550),interpolation=cv2.INTER_AREA)
cv2_imshow(input_image)
#print(input_image)
```

```python
gray_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
#print(gray_image)
```

```python
gray_image.shape
```

##SIFT

```python
sift = cv2.SIFT_create()
```

```python
kp = sift.detect(gray_image,None)
```

```python
img=cv2.drawKeypoints(gray_image,kp,input_image,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
cv2_imshow(img)
```

##ORB

```python
orb = cv2.ORB_create(5000)
```

```python
keypoints_orb, descriptors_orb = orb.detectAndCompute(gray_image, None)
```

```python
final_keypoints_orb = cv2.drawKeypoints(gray_image, keypoints_orb, input_image, (0,255,0))
```

```python
cv2_imshow(final_keypoints_orb)
```

##SURF

```python
surf = cv2.xfeatures2d.SURF_create(5000)
```

```python
kp, des = surf.detectAndCompute(input_image,None)
```

```python
len(kp)
```

```python
img2 = cv2.drawKeypoints(gray_image,kp,None,(255,0,0),4)
```

```python
cv2_imshow(img2)
```
